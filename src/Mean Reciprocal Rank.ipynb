```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
#temp = pd.read_csv('embeddings/250_0.16_vectors.tsv', sep=' ', header=None, skiprows=1)
temp = pd.read_csv("/h/224/stevenhl/chess_dim_skill/src/word2vecf/vectors/size_alpha/vecs_0.16_265.txt", sep=' ', header=None, skiprows=1)
temp = temp.drop(len(temp.columns)-1,axis=1)
problem_id = list(temp[0])
temp.rename(columns={0:'puzzle_id'}, inplace=True)
puzzle_id = list(temp['puzzle_id'])
temp.set_index('puzzle_id',inplace=True)
temp = temp.divide(np.linalg.norm(temp, axis=1), axis=0)
```

```python
temp
```

```python
problem_id
```

```python
#Sample tags and generate distance matrix of those tags
import sklearn.metrics
def generate_matrix(sampled_problems):
    vecs = temp.loc[sampled_problems]
    cos_sim = sklearn.metrics.pairwise.cosine_similarity(vecs)
    cos_dist = 1.0 - cos_sim
    #Set really small numbers to 0
    cos_dist=cos_dist.round(3)
    #Create cosine distance table
    table = pd.DataFrame(cos_dist, columns=sampled_problems,index=sampled_problems)
    return table
```

```python
table = generate_matrix(puzzle_id)
```

# <h2>MRR For Duplicate FEN</h2>

```python
duplicate_data = pd.read_csv("/w/225/1/chess/tactics/duplicate_FENs.csv")
```

```python
duplicate_data
```

```python
#Create a list of all problem IDs which have FEN duplicates with other IDs. The list is created for both ID1 and ID2
#the rank for ID2 wrt to ID1 might not be the same compared to the rank for ID1 wrt to ID2

duplicate_data_id = list(duplicate_data['ID1']) + list(duplicate_data['ID2'])
duplicate_data_id
```

```python
#Create a new table which is a modified version of table. The difference lies in the columns, where each column is
#the IDs in duplicate_data_id above. Basically, new_table is a "subset" of table. Note that the rows are still exactly
#the same as above since otherwise, the ranking won't be accurate.

new_table = pd.DataFrame(table, columns = duplicate_data_id)
new_table
```

```python
from tqdm.notebook import tqdm
tqdm().pandas()

r,c = new_table.shape

#temp is a matrix which is similar to new_table, except that each entry is a tuple of distance and its corresponding
#problem ID. The ordering for the tuple is (distance, problem_ID).
temp = np.empty([r,c], dtype = object)

for i in tqdm(range(r)):
    for j in range(c):
        temp[i][j] = (new_table.iloc[i,j], problem_id[i])
```

```python
temp
```

```python
#temp1 is the sorted version of temp. The sorting is done across each column.

temp1 = np.sort(temp, axis = 0)
temp1
```

```python
#Create a list that pairs one ID with its FEN duplicate ID in the form of tuple. Again, it's a 2-way pairing.

duplicate_id_to_id = []
r,c = duplicate_data.shape
for i in range(r):
    duplicate_id_to_id.append((duplicate_data.iloc[i,1], duplicate_data.iloc[i,2]))
    duplicate_id_to_id.append((duplicate_data.iloc[i,2], duplicate_data.iloc[i,1]))
```

```python
duplicate_id_to_id
```

```python
r,c = new_table.shape
reciprocal_rank = []
rank = []

for i in tqdm(range(len(duplicate_id_to_id))):
    current_id = duplicate_id_to_id[i][0] #get the ID on the left as the current_id
    target_id = duplicate_id_to_id[i][1]  #get the ID on the left as the target_id
    pos = duplicate_data_id.index(current_id) #find the column position
    if temp1[0][pos][0] != 'nan': #if a column consists of nan, then it's invalid
        for j in range(r):
            if target_id == temp1[j][pos][1]: #if we have found the desired duplicate ID
                rank.append(j+1) #get the rank (j+1 because indexing starts from 0, but rank starts from 1)
                reciprocal_rank.append(1./(j+1)) #get the reciprocal rank (j+1 because indexing starts from 0, but rank starts from 1)
```

```python
reciprocal_rank
```

```python
rank
```

```python
plt.hist(rank, bins = 1000)
```

```python
import matplotlib.pyplot as plt
plt.hist(reciprocal_rank, bins = 30)
```

```python
mean_reciprocal_rank = sum(reciprocal_rank) / len(reciprocal_rank)
mean_reciprocal_rank
```

# <h2> MRR for Duplicate FULL </h2>

```python
tactics_data = pd.read_csv('/w/225/1/chess/tactics/tactics_problem.csv')
tactics_data
```

```python
#a function to extract ONLY the full field if the full field exists. Otherwise, it's skipped
def extract_full(st):
    splitted_st = st.split("\\n") #splitted strings
    existence = 0
    for string in splitted_st:
        if "FULL" in string: #if we find the section which has FULL in it (as the indicator)
            target_st = string #take it as our target string
            existence = 1 #to denote that the FULL field indeed exists
            break
    if existence == 1: #if the FULL field exists, clean it up so that we get only the relevant information. Otherwise, skip
        target_st = target_st.replace('FULL', '')
        target_st = target_st.replace('[','')
        target_st = target_st.replace(']','')
        target_st = target_st.replace('"','')
        return target_st #return the cleaned target string if the FULL field exists
    return "N/A" #return N/A if the FULL field does not exist
```

```python
#Create a FULL column, append it to the tactics_data

from tqdm.notebook import tqdm
tqdm().pandas()

tactics_data['FULL'] = 'N/A'

row, col = tactics_data.shape
puzz_id = list(tactics_data['tactics_problem_id'])
for i in tqdm(range(row)):
    pos = puzz_id.index(tactics_data.iloc[i,0])
    pgn = tactics_data.iloc[pos,7]
    tactics_data.iloc[i,8] = extract_full(pgn)
```

```python
tactics_data
```

```python
#A list of all the cleaned FULL fields, EXCLUDING N/A's.

full_list_ori = np.array(tactics_data['FULL'])
full_list_new = []
for full in full_list_ori:
    if full != "N/A":
        full_list_new.append(full)
full_list_new
```

```python
#full_to_ids is a dictionary that maps FULL fields to ALL OF ITS CORRESPONDING PROBLEM IDs. 
full_to_ids = {}
for full in full_list_new:
    full_to_ids[full] = []
    
r,c = tactics_data.shape
for i in tqdm(range(r)):
    full = tactics_data.iloc[i,8]
    prob_id = tactics_data.iloc[i,0]
    if full != "N/A":
        full_to_ids[full].append(prob_id)
        
full_to_ids
```

```python
#duplicate_full_id_to_id is a list that pairs an ID to another FULL duplicate ID in the form of tuple. If there are more than two
#problems which are duplicate to each other, we form every possible 2-way pair of all the IDs.
duplicate_full_id_to_id = []
#target_problems is a list of all the problems which have at least one duplicate and so 
#we have to assess for their (reciprocal) rank
target_problems = []

for full in full_to_ids: #for every FULL field
    prob_id_list = full_to_ids[full] #get the list of corresponding problem IDs
    if len(prob_id_list) > 1: #if the list has duplicate IDs
        target_problems += prob_id_list #add that ID to the target_problems list
        #create all pairs of IDs in the list
        for k in range(len(prob_id_list)):
            for j in range(len(prob_id_list)):
                if prob_id_list[k] != prob_id_list[j]:
                    duplicate_full_id_to_id.append((prob_id_list[k],prob_id_list[j]))
                    
duplicate_full_id_to_id
```

```python
target_problems
```

```python
new_table_full = pd.DataFrame(table, columns = target_problems)
new_table_full
```

```python
r,c = new_table_full.shape

temp = np.empty([r,c], dtype = object)

for i in tqdm(range(r)):
    for j in range(c):
        temp[i][j] = (new_table_full.iloc[i,j], problem_id[i])
```

```python
temp
```

```python
temp1 = np.sort(temp, axis = 0)
temp1
```

```python
r,c = new_table_full.shape
reciprocal_rank_full = []
rank_full = []

for i in tqdm(range(len(duplicate_full_id_to_id))):
    current_id = duplicate_full_id_to_id[i][0] #get the ID on the left as the current
    target_id = duplicate_full_id_to_id[i][1] #get the ID on the right as the target
    pos = target_problems.index(current_id) #find the column position
    if temp1[0][pos][0] != 'nan': #if a column consists of nan, then it's invalid
        for j in range(r):
            if target_id == temp1[j][pos][1]:
                rank_full.append(j+1)
                reciprocal_rank_full.append(1./(j+1))
            
reciprocal_rank_full
```

```python
import matplotlib.pyplot as plt
plt.hist(reciprocal_rank_full, bins = 30)
```

```python
plt.hist(rank_full, bins = 1000)
```

```python
mean_reciprocal_rank_full = sum(reciprocal_rank_full) / len(reciprocal_rank_full)
mean_reciprocal_rank_full
```
